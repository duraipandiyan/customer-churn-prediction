### . Exploring Data ( Visualization )
####  Visualize age and Gender, City,  distribution to make some conclusion. 
- Histogram
- Kernel Distribution Function (KDE)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df=pd.read_csv('data/raw/ecommerce_customer_behavior_dataset_v1.csv')
df.head()
```

# Histogram plot to find the customers ages 

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 7))
plt.subplot(121)
sns.histplot(data=df,x='Age',bins=30,kde=True,color='g')
plt.subplot(122)
sns.histplot(data=df,x='Age',kde=True,hue='Gender')
plt.show()
```

#  Insights
1. The majority of the customer fall under age between 30 to 40
2. The maximum number of customers using the online platform are aged 18 t0 19

```python
# Columns city 
```

```python
fig, axs = plt.subplots(1, 2, figsize=(20, 7))
plt.subplot(121)
sns.histplot(data=df,x='City',bins=30,kde=True,color='g')
plt.subplot(122)
sns.histplot(data=df,x='City',kde=True,hue='Gender')
plt.show()
```

# Insights
1. Istanbul has the highest number of customers

2. Ankara, Izmir, Bursa, and Adana have a moderate customer base

3. Antalya, Gaziantep, Konya, Kayseri, and Eskisehir have relatively fewer customers

4. Females are high distribution compared to male

# Pie chart for gender percentages to understand their contributions.

```python
df['Gender'].value_counts()
plt.rcParams['figure.figsize']=(18,7)
plt.pie(x=df['Gender'].value_counts().values,labels=df['Gender'].value_counts().keys(),autopct='%1.2f%%')
```

# Insights
1. Female customers are the largest contributors to online platform purchases (50%).

2. Male customers closely follow, contributing 47.95% of total online purchases.

3. The Other category represents a very small portion of online platform customers (1.53%).

# Barplot for Age counts to understand their contributions 

```python
plt.rcParams['figure.figsize']=(18,7)
ax=sns.barplot(x=df['Age'].value_counts().index,y=df["Age"].value_counts().values)
for container in ax.containers:
    ax.bar_label(container,fmt='%.0f',padding=4)
```

# Insights

1. The maximum number of customers using the online platform are aged 18, with a count of around 1,600.

2. The majority of customers fall within the age range of 27 to 37, indicating this is the average and most active user group on the platform.

3. Customers aged 50 and above use the online platform in very low numbers compared to other age groups.

- Sessions with fewer than 3 page views are minimal, suggesting a low bounce rate.

- Very high engagement sessions (more than 14 page views) are rare and represent a small segment of users.

- Categories such as Electronics and Books show slightly higher page views, likely due to increased comparison before purchase.

- Female customers show slightly higher session activity compared to male customers.

# How many days does it take to delivey for the product?

```python
plt.subplots(1,3,figsize=(20,7))
plt.subplot(1,3,1)
ax=sns.barplot(x=df['Delivery_Time_Days'].value_counts().index,y=df['Delivery_Time_Days'].value_counts().values)
for container in ax.containers:
    ax.bar_label(container)
    
plt.subplot(1,3,2)
sns.histplot(data=df,x=df['Delivery_Time_Days'],hue='Product_Category',kde=True)

plt.subplot(1,3,3)
sns.histplot(data=df,x=df['Delivery_Time_Days'],hue='Gender',kde=True)
```

# Insights

1. The majority of customers received their products in less than 6 days.

2. Some customers received their products within 7 to 12 days.

3. A small number of customers received their products very quickly (within 12 days), while a few experienced longer delivery times of 13 to 25 days.

4. Delivery timelines do not vary significantly across product categories, as all categories follow a similar delivery pattern.

5. There is no notable difference in delivery time based on gender; both genders experience similar delivery timelines.
