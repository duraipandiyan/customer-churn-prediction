```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

customer_level=pd.read_csv('processed/customer_level_data.csv')
customer_level.head()
```

# Churn vs Recency

```python

plt.figure(figsize=(6,4))

ax = sns.barplot(
    data=customer_level,
    x="Churn",
    y="Recency_Days",
    estimator=np.mean
)

ax.set_title("Churn vs Recency")
ax.set_xlabel("Churn")
ax.set_ylabel("Average Recency (Days)")

plt.show()
```

# Churned customers show significantly higher recency, with an average gap of ~200 days since last purchase,  compared to ~40 days for retained customers.





# Churn vs Total_Orders and Churn vs Total_Spend

```python

fig,ax=plt.subplots(1,2,figsize=(18,7))
plt.subplot(121)


ax = sns.barplot(
    data=customer_level,
    x="Churn",
    y="Total_Orders",
    estimator=np.mean
)

ax.set_title("Churn vs Total_Orders")
ax.set_xlabel("Churn")
ax.set_ylabel("Total_Orders")



plt.subplot(122)

ax = sns.barplot(
    data=customer_level,
    x="Churn",
    y="Total_Spend",
    estimator=np.mean
)

ax.set_title("Churn vs Total_Spend")
ax.set_xlabel("Churn")
ax.set_ylabel("Total_Spend")



```

# Insight  Purchase Behavior:
. Churned customers exhibit significantly lower engagement, placing fewer orders and contributing lower total spend compared to retained customers.
# Actionable Angle:
. Early intervention strategies (discounts, reminders, loyalty offers) targeting low-frequency, low-spend customers could reduce churn risk.





# Churn vs Avg_Rating

```python

plt.figure(figsize=(6,4))

ax = sns.barplot(
    data=customer_level,
    x="Churn",
    y="Avg_Rating",
    estimator=np.mean
)

ax.set_title("Churn vs Avg_Rating")
ax.set_xlabel("Churn")
ax.set_ylabel("Avg_Rating")

plt.show()
```

# Insight  Customer Satisfaction:
 Average customer ratings are nearly identical for churned and retained customers, indicating that explicit satisfaction scores alone may not be a strong standalone indicator of churn.
#
# Interpretation:
 Customers may leave despite providing acceptable ratings, suggesting that churn is more strongly influenced by behavioral factors (recency, frequency, spend) than by self-reported satisfaction.

# Business Implication:
  Relying solely on customer ratings to predict churn could be misleading; behavioral engagement metrics should be prioritized for churn prediction.

# Key EDA Findings 

Recency is a strong indicator of churn

Lower order frequency and total spend are associated with churn

Customer ratings show minimal differentiation between churned and retained users
